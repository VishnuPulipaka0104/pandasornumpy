Get the Longest Album. This problem sounds a lot like #3 but the solution is quite a bit different. I would recommend looking up the SUM aggregate function.
Return the album name as Name, the album release year as Release Year, and the album length as Duration.

```python
new_table = albums.merge(songs,left_on="id",right_on="album_id",how="inner")
new_table = new_table[["name_x","release_year","length"]]
new_table = new_table.rename(columns={"name_x":"Name","release_year":"Release Year","length":"Duration"})
new_table = new_table.groupby(["Name","Release Year"])["Duration"].sum()
new_table = new_table.reset_index()
new_table.loc[new_table['Duration'] == new_table['Duration'].max()]
```

Select the longest Song off each Album

```python
new_songs = songs.merge(albums,left_on = "album_id",right_on = "id", how="inner")
new_songs = new_songs.rename(columns={"name_y":"Name", "release_year":"Release Year","length":"Duration"})
new_songs = new_songs[["Name","Release Year","Duration"]]
new_songs = new_songs.groupby(["Name","Release Year"])["Duration"].max()
new_songs
```

Get the number of Songs for each Band

```python
new_albums = albums.merge(songs,left_on="id",right_on="album_id",how="inner")
new_albums = new_albums.groupby("band_id")["name_y"].count()
new_bands = bands.merge(new_albums,left_on = "id", right_on = "band_id",how="inner")
new_bands = new_bands.rename(columns={"name":"Band","name_y":"Number of Songs"})
new_bands = new_bands[["Band","Number of Songs"]]
new_bands
```
